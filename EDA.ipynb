# Import libraries

```python
from EDA import plot_samples,plot_metric,compare_metrics
from utils import load_model
from torch import optim
from models import VAE
import torch
import pandas as pd
```

# Ignore warnings

```python
import warnings
warnings.filterwarnings("ignore")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

# VAE

## Load model

```python
vae_model = VAE(
                z_dim=10,
                beta=1,
                conditional=False,
                device=device)

vae_optimizer = optim.Adam(vae_model.parameters(), lr=1e-3)

vae_model, vae_optimizer = load_model(ckpt_path='./ckpts/ckpt_VAE_epoch200.ckpt',
                              model=vae_model,
                              optimizer=vae_optimizer)
```

## Reconstruct samples from test datas

```python
plot_samples(vae_model,
             model_name='VAE',
             conditional=False,
             classes_for_plot=[1,2,3],
             figsize=(12, 12),
             title_padding=0.93,
             title_fontsize=24,
             generate=False)
```

## Generate new samples by sampling from latent distribution

```python
plot_samples(vae_model,
             model_name='VAE',
             conditional=False,
             classes_for_plot=[1,2,3],
             figsize=(12, 12),
             title_padding=0.93,
             title_fontsize=24)
```

## Generate samples from latent by traversing over one dimension

```python
plot_samples(model=vae_model,
                 model_name='VAE',
                 conditional=False,
                 classes_for_plot=[1,2,3],
                 figsize=(12,12),
                 title_padding=0.93,
                 title_fontsize=24,
                 walk_over_dim='one',
                 variable_dim=6)
```

## Reconstruction-loss

```python
report_VAE = pd.read_csv('./reports/VAE_report.csv')
plot_metric(model_name='VAE',
            dataframe=report_VAE,
            metric_name='rec',
            figsize=(18,10),
            xylabel_fontsize=16,
            xy_ticks_fontsize=12,
            legend_fontsize=12,
            title_fontsize=20,
            linewidth_plot=4)
```

## KL-divergence

```python
plot_metric(model_name='VAE',
            dataframe=report_VAE,
            metric_name='kl',
            figsize=(18,10),
            xylabel_fontsize=16,
            xy_ticks_fontsize=12,
            legend_fontsize=12,
            title_fontsize=20,
            linewidth_plot=4)
```

## Negative-ELBO

```python
plot_metric(model_name='VAE',
            dataframe=report_VAE,
            metric_name='nelbo',
            figsize=(18,10),
            xylabel_fontsize=16,
            xy_ticks_fontsize=12,
            legend_fontsize=12,
            title_fontsize=20,
            linewidth_plot=4)
```
